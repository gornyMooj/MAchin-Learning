```python
import pandas as pd

# 1 solution
dzieci = pd.read_csv("DZIECI.txt", delimiter=';') 
przedszkola = pd.read_csv("PRZEDSZKOLA.txt", delimiter=';') 
preferencje = pd.read_csv("PREFERENCJE.txt", delimiter=';') 

dzieci.index = dzieci['Pesel']
przedszkola.index = przedszkola['Id_przedszkola']

dzieci = dzieci[(dzieci['Wiek'] == 6) & (dzieci['Plec'] == 'dziewczynka')]
przedszkola = przedszkola[przedszkola['Nazwa_przedszkola'].str.contains("75")]
przedzkole_id = list(przedszkola.index.values)[0]

preferencje = preferencje[preferencje['Id_przedszkola'] == przedzkole_id]
preferencje.index = preferencje['Pesel']
preferencje = preferencje['Id_przedszkola']
preferencje = dict(preferencje)
preferencje = list(preferencje.keys())

def znajdz(row):
    if row['Pesel'] in preferencje:
        return True
    else:
        return False
    

dzieci['preferencje'] = dzieci.apply(znajdz, axis=1)
dzieci = dzieci[dzieci['preferencje'] == True]
dzieci = dzieci.sort_values('Nazwisko')
dzieci
```

```python
# zlaczenia dataframes po indeksie
# 2nd solution


import pandas as pd
pd.set_option('display.max_columns', None)

# 2 solution
dzieci = pd.read_csv("DZIECI.txt", delimiter=';') 
przedszkola = pd.read_csv("PRZEDSZKOLA.txt", delimiter=';') 
preferencje = pd.read_csv("PREFERENCJE.txt", delimiter=';') 


df_all = preferencje.set_index('Pesel').join(dzieci.set_index('Pesel'))
df_all = df_all.set_index('Id_przedszkola').join(przedszkola.set_index('Id_przedszkola'))
display(df_all.head(2))

df_all[
    (df_all['Nazwa_przedszkola'] == 'Przedszkole nr 75') & 
    (df_all['Wiek'] == 6) & 
    (df_all['Plec'] == 'dziewczynka')].sort_values('Nazwisko')
```

```python
'''
zadanie 2
- wypisz wszystkie osoby ktore na liscie preferencji maja 3 przedszkola
'''
# solution 1

dzieci = pd.read_csv("DZIECI.txt", delimiter=';') 
przedszkola = pd.read_csv("PRZEDSZKOLA.txt", delimiter=';') 
preferencje = pd.read_csv("PREFERENCJE.txt", delimiter=';') 

df_all = preferencje.set_index('Pesel').join(dzieci.set_index('Pesel')).groupby(["Pesel", "Nazwisko", "Imie"]).size().reset_index(name='count')
df_all = df_all[df_all['count'] == 3]
df_all
```

```python
'''
zadanie 3
podaj nazwe przedszkola ktore zostalo wybrane jako pierwszy wybor
podaj nazwy tych osob
'''
dzieci = pd.read_csv("DZIECI.txt", delimiter=';') 
przedszkola = pd.read_csv("PRZEDSZKOLA.txt", delimiter=';') 
preferencje = pd.read_csv("PREFERENCJE.txt", delimiter=';') 

df_all = preferencje.set_index('Pesel').join(dzieci.set_index('Pesel'))
df_all = df_all.set_index('Id_przedszkola').join(przedszkola.set_index('Id_przedszkola'))

df_first = df_all[df_all['Numer_preferencji'] == 1]
df_second = df_all[df_all['Numer_preferencji'] == 2]
df_third = df_all[df_all['Numer_preferencji'] == 3]

merged = list(df_second['Nazwa_przedszkola'].unique()) + list(df_third['Nazwa_przedszkola'].unique())
df_first = df_first[~df_first['Nazwa_przedszkola'].isin(merged)]

print(df_first['Nazwa_przedszkola'].unique()[0])
print(df_first.shape[0])
df_first
```
