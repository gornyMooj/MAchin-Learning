```python
# od prowadzacego
# przygotowanie danych
x = [0,10,30,50, 100]
y = [2,4,7,5,12]
y1 = [-2,-14,-7,-15,-12]

# zaladowanie danych
plt.plot(x, y)
plt.plot(x, y1)

# wyswietlenie wyrenderowanego wykresu
plt.show()
```

```python
# numpy with matplotlib
import numpy as np

np_x = np.array([0,20,100])
np_y = np.array([0,20,100])

plt.plot(np_x, np_y, 'x') # dolozenie 'x' lub 'o' powoduje wyswietlenie zbior obserwacji

plt.show()
```

```python
linia_punktow = np.array([3,4,5,2,3,4,5,6])
plt.plot(linia_punktow)
plt.plot(linia_punktow, 'x')
plt.show()

# matplot dopowiedzial sobie ze na osi x mamy indexy w kolekcji
```

```python
# zamiana osi 
x = [0,10,30,50, 100]
y = [2,4,7,5,12]
plt.plot(y, x)

plt.show()
```

```python
# linewidth - grubosc linie

# marker - znacznik obserwacji  https://matplotlib.org/stable/api/markers_api.html

# linestyle - normalna linia
# linestyle : linia kropkowana
# linestyle -- kreskowana linia
# -. kreskowano kropkowana

# color
# r - red
# g - green
# b - blue
# c - cyan
# m - magenta
# y - yellow
# k - black
# w - white

linia_punktow = np.array([3,4,5,2,3,4,5,6])

plt.plot(linia_punktow, linewidth='1', marker='^', linestyle=':', color='pink')

plt.show()
```

```python
# ms - marker size
# mec - marker edge color 
# mfc = marker forground color 

plt.plot(x,y, marker='o', ms=25, mec='k', mfc='y')

plt.show()
```

```python
x1 = [0,4,5,6]
y1 = [4,5,6,8]
y2 = [1,2,3,1]

plt.plot(x,y, x1,y1, x1,y2)

plt.show()
```

```python
# font

# loc = 'center' , ' left', 'right'
f ={
    'family': 'Times New Roman',
    'color': 'blue',
    'size': 22
}

plt.grid()

plt.plot(x,y)
plt.xlabel('iksy')
plt.ylabel('igreki')

plt.title('Wykres z danymi', fontdict=f, loc='right')
plt.show()
```

```python
# axis can be x or y

plt.grid(axis='x', color='r', linestyle='-.', linewidth='0.3')

plt.plot(x,y)

plt.title('WYKRES Z DANYMI', fontdict=f, loc='right')
plt.show()
```

```python
# zadanie 1
# wygeneruj 3 listy po 10 elementow z zakresu od -10,10
# zaprezentuj na wykresie: r,g,b

from numpy import random

r=random.randint(-10,10, size=(30))
g=random.randint(-10,10, size=(30))
b=random.randint(-10,10, size=(30))

plt.plot(r, color='r', linestyle=':', marker='o')
plt.plot(g, color='g', linestyle='--', marker='D')
plt.plot(b, color='b', linestyle='-', marker='^')

plt.xlabel('numer losowania')
plt.ylabel('losowanie')

plt.legend(['r','g','b']) # legend

plt.show()
```

```python
# subplots 
from numpy import random

x1=random.randint(-10,10, size=(6))
y1=random.randint(-10,10, size=(6))

x2=random.randint(-10,10, size=(6))
y2=random.randint(-10,10, size=(6))

plt.subplot(1,2,1) # wymiar 1 kolumna x2 wiersze, ten jest pierwszy
plt.plot(x1,y1)

plt.subplot(1,2,2) # wymiar 1 kolumna x2 wiersze, ten jest drugi
plt.plot(x2,y2)

plt.show()
```

```python
# subplots 
from numpy import random

x1=random.randint(-10,10, size=(6))
y1=random.randint(-10,10, size=(6))

x2=random.randint(-10,10, size=(6))
y2=random.randint(-10,10, size=(6))

plt.subplot(1,2,1) # wymiar 1 wiersz x2 kolumna, ten jest pierwszy
plt.plot(x1,y1)
plt.title('SUBPLOTS 1')
plt.xlabel('sub1 iksy')
plt.ylabel('sub1 igreki')

plt.subplot(1,2,2) # wymiar 1 wiersz x2 wiersze, ten jest drugi
plt.plot(x2,y2)
plt.title('SUBPLOTS 2')
plt.xlabel('sub2 iksy')

plt.show()
```

```python
import pandas as pd

iris = pd.read_csv('iris.csv')
display(iris.head(2))
print( iris['species'].unique() )
print( iris['species'].value_counts() )
```

```python
colors = {'setosa':'tab:blue', 'versicolor':'tab:orange', 'virginica':'tab:green'}

plt.scatter( iris['petal_width'], iris['sepal_length'], c=iris['species'].map(colors))
plt.xlabel('petal_width')
plt.ylabel('sepal_length')
plt.title('1 scat')
plt.show()

plt.scatter( iris['sepal_length'], iris['sepal_width'], c=iris['species'].map(colors))
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.title('2 scat')
plt.show()

plt.scatter( iris['petal_width'], iris['petal_length'], c=iris['species'].map(colors))
plt.xlabel('petal_width')
plt.ylabel('petal_length')
plt.title('3 scat')
plt.show()

plt.scatter( iris['sepal_width'], iris['petal_width'], c=iris['species'].map(colors))
plt.xlabel('sepal_width')
plt.ylabel('petal_width')
plt.title('4 scat')
plt.show()
```

```python
colors = {'setosa':'tab:blue', 'versicolor':'tab:orange', 'virginica':'tab:green'}


plt.subplot(2,2,1) # 2 - liczba wierszy 2 - liczba kolumn 1- ktory w kolejnosci wykres
plt.scatter( iris['petal_width'], iris['sepal_length'], c=iris['species'].map(colors), alpha=0.5)
plt.xlabel('petal_width')
plt.ylabel('sepal_length')
plt.title('1 scat')


plt.subplot(2,2,2)  # 2 - liczba wierszy 2 - liczba kolumn 2- ktory w kolejnosci wykres
plt.scatter( iris['sepal_length'], iris['sepal_width'], c=iris['species'].map(colors), alpha=0.5)
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.title('2 scat')

plt.subplot(2,2,3)
plt.scatter( iris['petal_width'], iris['petal_length'], c=iris['species'].map(colors), alpha=0.5)
plt.xlabel('petal_width')
plt.ylabel('petal_length')
plt.title('3 scat')

plt.subplot(2,2,4)
plt.scatter( iris['sepal_width'], iris['petal_width'], c=iris['species'].map(colors), alpha=0.5)
plt.xlabel('sepal_width')
plt.ylabel('petal_width')
plt.title('4 scat')

plt.show()
```

```python
# zad wykres liniowy
# czerwona obwodka najnizszy kurs
# zielona obwodka najwyzszy kurs
# markery jako kolka
# linia dla kupna i linia dla sprzedazy

import pandas as pd

df = pd.read_csv('kursy-EUR.csv', delimiter=';')
display(df.head(2))

df['Data'] = df['Data'].astype("datetime64")
df['Kupno'] = df['Kupno'].str.replace(',', '.').astype('float')

df['Sprzeda'] = df['Sprzeda'].str.replace(',', '.').astype('float')

display(df.info())

df['kupno_kolors'] = 'blue'
df.loc[df['Kupno'] == df['Kupno'].min(), 'kupno_kolors'] = 'red'
df.loc[df['Kupno'] == df['Kupno'].max(), 'kupno_kolors'] = 'green'

colors = {'blue':'blue', 'red':'red', 'green':'green'}

plt.plot(df['Data'],df['Kupno'], linestyle='--', color='blue')
plt.scatter(df['Data'],df['Kupno'], c=df['kupno_kolors'], s=60 )
plt.scatter(df['Data'],df['Kupno'], c='blue', s=30 )


df['sprzedaz_kolors'] = 'pink'
df.loc[df['Sprzeda'] == df['Sprzeda'].min(), 'sprzedaz_kolors'] = 'red'
df.loc[df['Sprzeda'] == df['Sprzeda'].max(), 'sprzedaz_kolors'] = 'green'

plt.plot(df['Data'],df['Sprzeda'], linestyle='--', color='pink')
plt.scatter(df['Data'],df['Sprzeda'], c=df['sprzedaz_kolors'], s=60 )
plt.scatter(df['Data'],df['Sprzeda'], c='pink', s=30 )



plt.xticks(rotation=90)
plt.xticks(color='r')
plt.legend(['Kupno', 'Sprzeda'])
plt.show()
```
