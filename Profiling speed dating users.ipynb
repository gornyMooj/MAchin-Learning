# Algorytm K średnich

```python
from sklearn.cluster import KMeans #import biblioteki klastrowej Kmeans
from sklearn.preprocessing import StandardScaler #standaryzowanie cech aby mieć pewność że uczenie maszynowe na pewno zadziała
import numpy as np #import biblioteki numpy w tym kodzie przydatnej do pracy na macierzach
```

```python
scaler = StandardScaler() #tworzenie własnego skalera
data_subset_scaled = scaler.fit_transform(data_subset) #tworzenie zmiennej dataset przeskalowanej
data_subset_scaled
```

```python
data_subset_scaled.mean(axis=0) #standaryzacja do zera
```

```python
print(data_subset.to_string()) #przedstawienie datasetu jako string
```

```python
data_subset_scaled.dropna(inplace=True) 
```

```python
data_subset.dropna(inplace=True)
data_subset_scaled = scaler.fit_transform(data_subset)
data_subset_scaled.mean(axis=0) #You can also check var
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=200, n_init=20, random_state=0)
labels = kmeans.fit_predict(data_subset_scaled) #wybranie 3 klastrów
```

```python
labels #wyświetlenie wartości tablicy, które muszą być od 0 do 2
```

```python
data_subset
```

# Analiza PCA

```python
from sklearn.decomposition import PCA #import biblioteki od PCA
pca = PCA(n_components=2) 
data_subset_pca = pca.fit_transform(data_subset_scaled) #tworzenie datasetu do analizy PCA
data_subset_pca
```

```python
data_subset_tf = pd.concat([pd.DataFrame(data_subset_pca, columns=['PC1', 'PC2']), 
                           pd.DataFrame(labels, columns=['labels'])], axis=1) #concat służy łączeniu najczęściej się używa join inner lub outer
data_subset_tf
```

```python
plot = data_subset_tf.plot.scatter(x='PC1', y='PC2', c=data_subset_tf.labels.replace([0, 1, 2, 3], ['b', 'r', 'g', 'y']))
fig = plot.get_figure()
fig.savefig("wynik.png", dpi=600) #wyświetlenie wyniku analizy PCA
```

```python
print(pca.components_)
```

```python
import numpy as np
np.matmul(pca.components_, np.array([10, 10, 0, 70, 10, 0])) 
```

```python
import matplotlib.pyplot as plt
data_subset_tf.plot.scatter(x='PC1', y='PC2', c=data_subset_tf.labels.replace([0, 1, 2, 3], ['b', 'y', 'g', 'r']), figsize=(10,10))
for i, colname in enumerate(data_subset.columns):
    plt.annotate(colname, ha='center', va='bottom', xy=(0, 0), size=15,
             xytext=(pca.components_[0,i]*4, pca.components_[1,i]*4), 
             arrowprops = {'arrowstyle':'<-'}) #wyświetlenie analizy PCA wraz z pokazaniem odchylenia cech
```
